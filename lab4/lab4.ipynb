```python
"""
Task 1 Plotting Real Data
"""

import pandas as pd
import matplotlib.pyplot as plt

# define figure parameters
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# read in csv file
headers = ['TIME', 'CH1', 'CH2']
f = pd.read_csv(('/content/drive/MyDrive/e12/lab4/data/part1_better.csv'),
                skiprows=21, names = headers)

# define start and end points in data (30s)
start = f.CH1[f.CH1 > 0.03].index[0] - 22
end = f.TIME[f.TIME > (30 + f.TIME[start])].index[0]

# plot data, displaced to start at 0s
plt.plot(f.TIME[start:end] - f.TIME[start], f.CH1[start:end], markersize = "0.01")
plt.plot(f.TIME[start:end] - f.TIME[start], -f.CH2[start:end], markersize = "0.01")
plt.title("Task 1 Data Force Across Sensors")
plt.xlabel("Time (sec)")
plt.ylabel("Force (N)")
```

```python
"""
Task 1 Plotting ODE solution
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

u = 0.030 # define system vars
k_c = 53
theta_1_0 = 10 * np.pi/180 # initial conditions
theta_2_0 = 10 * np.pi/180
lambda_1 = 25.422 # lambdas
lambda_2 = 11.422
c_1 = 0.5*theta_1_0 + 0.5*theta_2_0 # constants
c_2 = 0.5*theta_1_0 - 0.5*theta_2_0

# define found solutions for theta_1 and theta_2
def theta_1(t):
  return c_1*np.cos(np.sqrt(lambda_1) * t) + c_2 * np.cos(np.sqrt(lambda_2) * t)

def theta_2(t):
  return c_1*np.cos(np.sqrt(lambda_1) * t) - c_2 * np.cos(np.sqrt(lambda_2) * t)

# define figure parameters
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# plot force across sensors
space = np.linspace(0, 30, 1000)
f_1 = u*k_c*theta_1(space)
plt.plot(space, f_1)
f_2 = u*k_c*theta_2(space)
plt.plot(space, f_2)
plt.title("Task 1 ODE Solution Force across Sensors")
plt.xlabel("Time (sec)")
plt.ylabel("Force (N)")
```
